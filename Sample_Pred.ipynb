Support Vector Machine (SVM)

```python
import joblib

# Loading the SVM model and TF-IDF vectorizer with n-grams
svm_model = joblib.load('/content/drive/MyDrive/NLP_Coursework/svm_model.pkl')
tfidf_vectorizer = joblib.load('/content/drive/MyDrive/NLP_Coursework/tfidf_ngram.pkl')
```

```python
# Sample prediction
sample_text = ["Vaccines are approved by WHO and safe to use."]
X_sample = tfidf_vectorizer.transform(sample_text)
svm_pred = svm_model.predict(X_sample)

print("SVM Prediction:", svm_pred[0])  # 0 = Real, 1 = Fake
```

Bidirectional LSTM (BiLSTM) with GloVe (Trainable)

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.sequence import pad_sequences
import pickle

# Loading the BiLSTM model
bilstm_model = load_model('/content/drive/MyDrive/NLP_Coursework/bilstm_glove_trainable.h5')

# Loading the tokenizer
with open('/content/drive/MyDrive/NLP_Coursework/tokenizer.pkl', 'rb') as f:
    tokenizer = pickle.load(f)

# Loading max_len
with open('/content/drive/MyDrive/NLP_Coursework/max_len.txt', 'r') as f:
    max_len = int(f.read())
```

```python
# Sample prediction
sample_text = ["This article falsely claims the Earth is flat."]
seq = tokenizer.texts_to_sequences(sample_text)
padded_seq = pad_sequences(seq, maxlen=max_len, padding='post')

bilstm_pred = bilstm_model.predict(padded_seq)
print("BiLSTM Prediction (probability):", bilstm_pred[0][0])
print("Predicted class:", int(bilstm_pred[0][0] > 0.5))  # 0 = Real, 1 = Fake
```
